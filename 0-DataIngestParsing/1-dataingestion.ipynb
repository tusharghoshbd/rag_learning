### TextLoader - Read single file


```python
from langchain_community.document_loaders import TextLoader

loader  = TextLoader("data/text_files/python_intro.txt", encoding='utf-8')

document = loader.load()

# print(document)
print(f" content : {document[0].page_content[0:100]}....")
```

```python

from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
        "data/text_files", 
        glob="**/*.txt", 
        loader_cls=TextLoader,
        loader_kwargs= { 'encoding': "utf-8"},
        show_progress= True
    )

documents = dir_loader.load()
print(f"loader {len(documents)}")
documents

for i, doc in enumerate(documents):
    print("Document", i)
    print ("Content", doc.page_content)
    print ("Metadata soruce", doc.metadata['source'])

```

```python

from langchain_text_splitters import (CharacterTextSplitter, RecursiveCharacterTextSplitter, TokenTextSplitter)

char_splitter = CharacterTextSplitter(
    separator= "\n",
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len
)

char_chunks = char_splitter.split_text(documents[0].page_content)

print(len(char_chunks))
```

```python
## Recursive character splitter

recursive_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""],  # order matters
    chunk_size=200,
    chunk_overlap=20
)

# Split text
chunks = recursive_splitter.split_text(documents[0].page_content)
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:")
    print(chunk)
    print("-" * 30)
```

```python
# Token Text splitter

# Create token-based splitter
splitter = TokenTextSplitter(
    chunk_size=200,      # max tokens per chunk
    chunk_overlap=20     # overlap in tokens
)

# Split text
chunks = splitter.split_text(documents[0].page_content)

# Print result
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}")
```
